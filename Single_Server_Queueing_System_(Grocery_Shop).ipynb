```python
import random

def main():
    print("Simulation for A Small Grocery Store")

    total_customers = int(input("Enter the total number of customers(between 0 and 6): "))

    iat, st, at, tsb, tse, wt, tss, ios = [0] * total_customers, [0] * total_customers, [0] * total_customers, [0] * total_customers, [0] * total_customers, [0] * total_customers, [0] * total_customers, [0] * total_customers

    random.seed()

    print("\nEnter random digits for {} customers arrival(between 1 and 1000):".format(total_customers))
    for i in range(total_customers):
        at_input = int(input("Random Digit for Customer {} arrival: ".format(i)))

        if at_input < 126:
            iat[i] = 1
        elif at_input < 251:
            iat[i] = 2
        elif at_input < 376:
            iat[i] = 3
        elif at_input < 501:
            iat[i] = 4
        elif at_input < 626:
            iat[i] = 5
        elif at_input < 751:
            iat[i] = 6
        elif at_input < 876:
            iat[i] = 7
        else:
            iat[i] = 8

        at[i] = 0 if i == 0 else at[i - 1] + iat[i]

        print("Customer {}: Arrival Time={} Inter-Arrival Time={}".format(i, at[i], iat[i]))

    print("\nEnter random digits for {} customers service time(between 1 and 1000):".format(total_customers))
    for i in range(total_customers):
        st_input = int(input("Random Digit for Customer {} service time: ".format(i)))

        if st_input < 11:
            st[i] = 1
        elif st_input < 31:
            st[i] = 2
        elif st_input < 61:
            st[i] = 3
        elif st_input < 86:
            st[i] = 4
        elif st_input < 96:
            st[i] = 5
        elif st_input < 101:
            st[i] = 6

        print("Customer {}: Service Time={}".format(i, st[i]))

    print("\n")

    avg_waiting_time = 0
    probability_to_wait = 0
    avg_service_time = 0
    avg_interarrival_time = 0
    avg_waiting_time_for_waiters = 0
    avg_time_in_system = 0
    total_idle_time = 0
    num_customers_waited = 0

    for i in range(total_customers):
        if i == 0:
            tsb[i] = at[i]
        else:
            tsb[i] = at[i] if at[i] >= tse[i - 1] else tse[i - 1]

        tse[i] = tsb[i] + st[i]
        wt[i] = tsb[i] - at[i]
        tss[i] = st[i] + wt[i]
        ios[i] = tsb[i] if i == 0 else tsb[i] - tse[i - 1] if tsb[i] > tse[i - 1] else 0
        total_idle_time += ios[i]

        avg_waiting_time += wt[i]
        avg_service_time += st[i]
        avg_interarrival_time += iat[i]
        avg_time_in_system += tss[i]

        if wt[i] > 0:
            probability_to_wait += 1
            avg_waiting_time_for_waiters += wt[i]
            num_customers_waited += 1

        print("Customer {}: tsb={} wt={} tse={} tss={} ios={}".format(i, tsb[i], wt[i], tse[i], tss[i], ios[i]))

    avg_waiting_time /= total_customers
    probability_to_wait /= total_customers
    avg_service_time /= total_customers
    avg_interarrival_time /= (total_customers - 1)
    avg_waiting_time_for_waiters /= num_customers_waited if num_customers_waited > 0 else 1
    avg_time_in_system /= total_customers

    total_simulation_time = tse[total_customers - 1]
    probability_of_idle_time = total_idle_time / total_simulation_time

    print("\nAverage Waiting Time: {:.3f}".format(avg_waiting_time))
    print("Probability Customer Has to Wait in Queue: {:.3f}".format(probability_to_wait))
    print("Probability of Idle Time: {:.3f}".format(probability_of_idle_time))
    print("Average Service Time: {:.3f}".format(avg_service_time))
    print("Average Inter-Arrival Time: {:.3f}".format(avg_interarrival_time))
    print("Average Waiting Time for Those Who Wait: {:.3f}".format(avg_waiting_time_for_waiters))
    print("Average Time Customers Spent in System: {:.3f}".format(avg_time_in_system))

if __name__ == "__main__":
    main()
```
